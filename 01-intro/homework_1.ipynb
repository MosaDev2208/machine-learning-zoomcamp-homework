```python
# Question 3

# Check the unique values in the 'fuel_type' column
unique_fuel_types = df['fuel_type'].unique()
print("Unique fuel types:", unique_fuel_types)

# Count how many unique ones there are
num_fuel_types = df['fuel_type'].nunique()
print("Number of unique fuel types:", num_fuel_types)
```

```python
# Question 4

# Count missing values in each column
missing_counts = df.isna().sum()
print("Missing values per column:\n", missing_counts)

# Count columns with any missing values
num_missing_columns = (missing_counts > 0).sum()
print("\nNumber of columns with missing values:", num_missing_columns)

# (Optional) List the columns with missing values
missing_columns = missing_counts[missing_counts > 0].index.tolist()
print("Columns with missing values:", missing_columns)
```

```python
# Question 5

# Check unique origins to confirm 'Asia' is present
print("Unique origins:", df['origin'].unique())

# Filter for Asia and find max fuel efficiency
asia_cars = df[df['origin'] == 'Asia']  # Select rows where origin is 'Asia'
max_efficiency = asia_cars['fuel_efficiency_mpg'].max()  # Get the maximum MPG
print("Maximum fuel efficiency for Asia cars:", max_efficiency)
```

```python
# Question 6

# Step 1: Median horsepower before filling
median_hp_before = df['horsepower'].median()
print("Median horsepower before filling:", median_hp_before)

# Step 2: Most frequent (mode) horsepower value
mode_hp = df['horsepower'].mode()[0]  # mode() returns a Series; [0] gets the first (most frequent)
print("Most frequent horsepower value:", mode_hp)

# Step 3: Fill missing values with the mode
df['horsepower_filled'] = df['horsepower'].fillna(mode_hp)  # Creates a new column with filled values

# Step 4: Median horsepower after filling
median_hp_after = df['horsepower_filled'].median()
print("Median horsepower after filling:", median_hp_after)

# Check if it changed
if median_hp_after > median_hp_before:
    print("Yes, it increased")
elif median_hp_after < median_hp_before:
    print("Yes, it decreased")
else:
    print("No")
```

```python
# Question 7

# 1. Select all cars from Asia

asia_cars = df[df['origin'] == 'Asia']
asia_cars
```

```python
# 2. Select only 'vehicle_weight' and 'model_year' columns

asia_cars_selected = asia_cars[['vehicle_weight', 'model_year']]
asia_cars_selected.head()
```

```python
# 3. Select the first 7 values

asia_cars_7 = asia_cars_selected.head(7)
asia_cars_7 
```

```python
# 4. Extract and display the underlying NumPy array from the DataFrame

X = asia_cars_7 .to_numpy()
X
```

```python
# 5. Compute the matrix multiplication of the transpose of X and X. Call the result XTX

XTX = X.T @ X
XTX
```

```python
# 6. Invert XTX

XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
# 7. Create an array y with the following values: 1100, 1300, 800, 900, 1000, 1100, 1200.

y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y
```

```python
# 8. Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w

w = XTX_inv @ X.T @ y
w
```

```python
# 9. What's the sum of all the elements of the result?

result_sum = np.sum(w)
result_sum
print("The sum of the elements of w is:", result_sum)
```
